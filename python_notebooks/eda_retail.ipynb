```python
# EDA

# total sales and profit year wise
df.groupby('year')[['sales','profit']].sum().plot(kind='bar',figsize=(8,4))

# insights >> 2014 has highes sales with maximum profit, basically as the sales increases , profit also increases
```

```python
# category wise total sales and profit
df.groupby('category')[['sales','profit']].sum().sort_values('sales', ascending=False)
# insights >> technology is having highest sales and profit
```

```python
# division of profit category wise
df.groupby('category')['profit'].sum().plot.pie(autopct='%1.1f%%')  
plt.title('Profit Share by Category')
plt.show()

#insights >> technology is having maximum profit
```

```python
df.groupby('region')['profit_margin'].mean()
```

```python
# discount impact on profit
sns.scatterplot(x='discount', y='profit', data=df)
# insights >> as the discount increases >> profit decreases
```

```python
# top selling products
top_products = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)
```

```python
top_products
# insights >> Apple Smart Phone, Full Size has the highest sales
```

```python
# region wise sales and profit
region_profit = df.groupby('region')[['sales','profit']].sum().sort_values('profit', ascending=False)
region_profit.plot(kind='bar', figsize=(8,4))
plt.title('Sales and Profit by Region')
plt.ylabel('Amount')
plt.show()

# insights >> central region has highest sales and profit , profit is comparatively lower than sales in every region
# lowest sales in cananda 
```

```python
# monthly sales trend
monthly = df.groupby(['year','month'])['sales'].sum().reset_index()
monthly['month_year'] = monthly['year'].astype(str) + '-' + monthly['month'].astype(str)
plt.figure(figsize=(12,4))
sns.lineplot(x='month_year', y='sales', data=monthly, marker='o')
plt.xticks(rotation=60)
plt.title('Monthly Sales Trend')
plt.show()

# insights 
# the sales significantly increases from year 2011 to 2014
#there is always a growth in the sales in 4 quarter of every year
```

```python
# top customers by profit 
top_customers = df.groupby('customer_name')['profit'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='barh', figsize=(8,4))
plt.title('Top 10 Customers by Profit')
plt.xlabel('Profit')
plt.show()
```

```python
# profit margin distribution
sns.histplot(df['profit_margin'], bins=30, kde=True)
plt.title('Distribution of Profit Margins')
plt.show()

# insights >> profit margin is mostly nearly negative and zero that means loss
# there is rarely hight profit margin
```

```python
# how disount affectng the profit
discount_profit = df.groupby('discount_band')['profit'].mean().reset_index()
sns.barplot(x='discount_band', y='profit', data=discount_profit)
plt.title('Average Profit by Discount Band')
plt.show()

# insights >> lower discount > high profit
# higher discount > low profit
```

```python
sns.scatterplot(x='quantity', y='profit', hue='category', data=df)
plt.title('Profit vs Quantity Sold by Category')
plt.show()

#insights
# > High quantity ≠ High profit
# > Category selection = Profit impact
# > Furniture is risky business
# > Technology best return deti hai, par inconsistent
```

```python
# shipping cost vs profit
sns.scatterplot(x='shipping_cost', y='profit', data=df)
plt.title('Shipping Cost vs Profit')
plt.show()
#Higher shipping costs are generally associated with lower profits
```

```python
sns.heatmap(df[['sales','profit','discount','quantity','shipping_cost','profit_margin']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# insights >>
# > When sales increase, profits also increase — but not by much, which means the profit model is weak.
# > Giving discounts kills profit — this is a clear issue.
# > Shipping cost rises with sales, but profit doesn’t rise enough — delivery is too expensive.
```

```python
engine.execute("""
CREATE TABLE daily_kpi AS
SELECT
  order_date::date AS order_date,
  SUM(sales) AS total_sales,
  SUM(profit) AS total_profit,
  ROUND(100.0 * SUM(profit)/SUM(sales),2) AS profit_margin,
  COUNT(DISTINCT customer_id) AS active_customers
FROM sales_raw
GROUP BY order_date
ORDER BY order_date;
""")
```
